### PCA on the iris dataset

```python
iris = datasets.load_iris()
X = iris.data
X
```

```python
variable_names = iris.feature_names
X[0:10,]
```

```python
pca =decomposition.PCA()
iris_pca = pca.fit_transform(X)

pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

The explained variance ration tells us how much information is compressed into the first few components.
When you are deciding how many components to keep, look at the percent of cumulative variance. Make sure to retain
at least 70% of the dataset's original information.

```python
comps = pd.DataFrame(pca.components_, columns=variable_names)
comps
```

```python
sb.heatmap(comps)
```

```python
The heatmap shows that:
    1. Principal component one (0) is strongly positively correlated with petal length, and moderately positively corelated
    with sepal length and petal width, and slightly negatively correlated sepal width.
    2. Principal component two (1) is stronly negatively correlated with petal length and petal width and strongly correlated
    with sepal length and sepal width
```
